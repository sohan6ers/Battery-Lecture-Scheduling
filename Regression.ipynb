```python
#https://www.w3schools.com/python/python_ml_polynomial_regression.asp
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import numpy
fqd = pd.read_csv('FinalQualityDataCSVFile.csv')
```

```python
#x_dataset
x_dataset = [i for i in range(0,24)]
```

```python
fqd['series_name'].unique()
```

```python
#y_dataset
# Building 0
b0_dataset = []
for i in range(0,24):
    temp_dataset = fqd.loc[(fqd['series_name'] == 'Building0') & (fqd['Hour'] == i)]
    average = temp_dataset["series_value"].mean()
    b0_dataset.append(average)
    
b0_regression = numpy.poly1d(numpy.polyfit(x_dataset, b0_dataset, 5))
the_line_drawn = numpy.linspace(1, 22, 100)
plt.scatter(x_dataset, b0_dataset)
plt.plot(the_line_drawn, b0_regression(the_line_drawn))
```
